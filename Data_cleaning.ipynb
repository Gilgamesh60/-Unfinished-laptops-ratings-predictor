# Cleaning the review data so that we can perform sentiment analysis afterwards

```python
! pip install -q wordcloud
import wordcloud
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words= stopwords.words('english')
```

```python
##cleaning reviews for sentiment analysis
#(1)lowering the capital letters
#(2)replacing all the punctuation marks
#(3)deleting all the stop words
for i in range(5):
  df.replace({f'Review{i+1}':{np.nan:''}},inplace=True)
  df[f'Review{i+1}']=df[f'Review{i+1}'].apply(lambda x: " ".join(word.lower() for word in x.split()))
  df[f'Review{i+1}']=df[f'Review{i+1}'].str.replace('[^\w\s]','')
  df[f'Review{i+1}']=df[f'Review{i+1}'].apply(lambda x:" ".join(word for word in x.split() if word not in stop_words ))
df
```

#Cleaning description for sentiment analysis

```python
df['Description']
```

```python
#converting NaN elements into string
df.replace({'Description':{np.nan:''}},inplace=True)
```

```python
# Removing Punctuations from the Review description
import string
def punctuation_removal(messy_str):
    clean_list = [char for char in messy_str if char not in string.punctuation]
    clean_str = ''.join(clean_list)
    return clean_str

df['Description'] = df['Description'].apply(punctuation_removal)
```

```python
# lets make a function to remove Numbers from the reviews
import re
def drop_numbers(list_text):
    list_text_new = []
    for i in list_text:
        if not re.search('\d', i):
            list_text_new.append(i)
    return ''.join(list_text_new)

df['Description'] = df['Description'].apply(drop_numbers)
```

```python
# lets create a function to remove accented characters
import unicodedata
def remove_accented_chars(text):
    new_text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    return new_text

# lets apply the function
df['Description'] = df.apply(lambda x: remove_accented_chars(x['Description']), axis = 1)
```

```python
# Create a function to remove special characters
def remove_special_characters(text):
    pat = r'[^a-zA-z0-9]' 
    return re.sub(pat, ' ', text)
 
# lets apply this function
df['Description'] = df.apply(lambda x: remove_special_characters(x['Description']), axis = 1)
```

```python
#lower all the capital characters and remove the 'read more' part at the end of few strings which is not necessary
df['Description']=df['Description'].apply(lambda x: " ".join(word.lower() for word in x.split()))
df['Description']=df['Description'].apply(lambda x: x.split('read more')[0])
```
