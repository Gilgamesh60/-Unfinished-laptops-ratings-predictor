**We will take summary of top 5 reviews and entire description of the most helpful review for performing sentiment analysis afterwards**

```python
BASE_URL='https://www.flipkart.com/'
def Reviewscrapper(url):
  import numpy as np
  driver.get(url)
  a=driver.find_elements(by=By.CLASS_NAME,value='_1fQZEK')
  b=[]
  c=[]
  d=[]
  e=[]
  for i in range(len(a)):
      b.append(a[i].get_attribute('outerHTML'))
      c.append(b[i].split('href="')[1])
      d.append(c[i].split('">')[0])
      e.append(d[i].split('&q=laptops')[0])
  product_url=[]
  parsed_url=[]
  for i in range(len(d)):
     product_url.append(urljoin(BASE_URL,e[i]))
     parsed_url.append(urlparse.urlparse(product_url[i]))
     parsed_url_path = parsed_url[i].path
     parsed_url_path_split = parsed_url_path.split("/")
     parsed_url_path_split[2] = "product-reviews"
     parsed_url_path_modified = "/".join(parsed_url_path_split)
     parsed_url_modified = parsed_url[i]._replace(path=parsed_url_path_modified)
     product_url[i]=parsed_url_modified.geturl()
  temp_review1=[]
  temp_review2=[]
  temp_review3=[]
  temp_review4=[]
  temp_review5=[]
  temp_description=[]
  for i in range(len(product_url)):
    driver.get(product_url[i])
    reviews=driver.find_elements(by=By.CLASS_NAME,value='_2-N8zT')
    
    try:
      temp_review1.append(reviews[0].text)
    except:
      temp_review1.append(np.nan)
    try:
      temp_review2.append(reviews[1].text)
    except:
      temp_review2.append(np.nan)
    try:
      temp_review3.append(reviews[2].text)
    except:
      temp_review3.append(np.nan)
    try:
      temp_review4.append(reviews[3].text)
    except:
      temp_review4.append(np.nan)
    try:
      temp_review5.append(reviews[4].text)
    except:
      temp_review5.append(np.nan)

    reviews=driver.find_elements(by=By.CLASS_NAME,value='t-ZTKy')
    try:
       temp_description.append(reviews[0].text)
    except:
       temp_description.append(np.nan)


   
  return [temp_review1,temp_review2,temp_review3,temp_review4,temp_review5,temp_description]
```

```python
review1=[]
review2=[]
review3=[]
review4=[]
review5=[]
description=[]
displaylength=[]
no_of_pages=39
for i in range(no_of_pages):
  review1.extend(Reviewscrapper(f'{BASE_URL}search?q=laptops&sort=price_asc&page={i+1}')[0])
  review2.extend(Reviewscrapper(f'{BASE_URL}search?q=laptops&sort=price_asc&page={i+1}')[1])
  review3.extend(Reviewscrapper(f'{BASE_URL}search?q=laptops&sort=price_asc&page={i+1}')[2])
  review4.extend(Reviewscrapper(f'{BASE_URL}search?q=laptops&sort=price_asc&page={i+1}')[3])
  review5.extend(Reviewscrapper(f'{BASE_URL}search?q=laptops&sort=price_asc&page={i+1}')[4])
  description.extend(Reviewscrapper(f'{BASE_URL}search?q=laptops&sort=price_asc&page={i+1}')[5])
```

```python
BASE_URL='https://www.flipkart.com/'
def reviews(url):
  import numpy as np
  driver.get(url)
  a=driver.find_elements(by=By.CLASS_NAME,value='_1fQZEK')
  b=[]
  c=[]
  d=[]
  for i in range(len(a)):
      b.append(a[i].get_attribute('outerHTML'))
      c.append(b[i].split('href="')[1])
      d.append(c[i].split('">')[0])
  product_url=[]
  parsed_url=[]
  for i in range(len(d)):
     product_url.append(urljoin(BASE_URL,d[i]))
     parsed_url.append(urlparse.urlparse(product_url[i]))
     parsed_url_path = parsed_url[i].path
     parsed_url_path_split = parsed_url_path.split("/")
     parsed_url_path_split[2] = "product-reviews"
     parsed_url_path_modified = "/".join(parsed_url_path_split)
     parsed_url_modified = parsed_url[i]._replace(path=parsed_url_path_modified)
     product_url[i]=parsed_url_modified.geturl()
  
  p=[]
  t=[]
  for i in range(len(product_url)):
    driver.get(product_url[i])
    try:
      r=driver.find_elements(by=By.CLASS_NAME,value='_3LWZlK')
      p.append(r[0].text)
    except:
      p.append(np.nan) 
    try:
      d=driver.find_elements(by=By.CLASS_NAME,value='_2_R_DZ')
      t.append(d[0].text)
    except:
      t.append(np.nan)
   
  return [p,t]
```
